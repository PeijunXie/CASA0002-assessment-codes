```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
from operator import itemgetter
G = nx.read_graphml('london.graphml')
type(G)
```

```python
#Degree Centrality
```

```python
deg_london =nx.degree_centrality(G)
nx.set_node_attributes(G,dict(deg_london),'degree')
# To dataframe using the nodes as the index
df = pd.DataFrame(index=G.nodes())
#df['line_name'] = pd.Series(nx.get_node_attributes(G, 'line_name'))
df['degree'] = pd.Series(nx.get_node_attributes(G, 'degree'))

df_sorted = df.sort_values(["degree"], ascending=False)
df_sorted[0:10]
```

```python
G2A=G.copy()
l2A=list(G2A.nodes)
len(l2A)
```

```python
deg_london =nx.degree_centrality(G2A)
nx.set_node_attributes(G2A,dict(deg_london),'degree')
# To dataframe using the nodes as the index
df = pd.DataFrame(index=G2A.nodes())
#df['line_name'] = pd.Series(nx.get_node_attributes(G, 'line_name'))
df['degree'] = pd.Series(nx.get_node_attributes(G2A, 'degree'))

df_sorted = df.sort_values(["degree"], ascending=False)
df_sorted[0:10]
```

```python
#strategy A
a = 1;
while a<11:

    
    degree_values=[(i[1]['degree']) for i in G2A.nodes(data=True)]
    maxbet=[n for n in G2A.nodes() if G2A.nodes[n]['degree'] == max(degree_values)]
    c=list()
    c.append(maxbet[0])
    print(c)
    G2A.remove_nodes_from(c)

    global_eff=nx.global_efficiency(G2A)
    global_eff
    print(global_eff)

    cluster_coe=nx.average_clustering(G2A)
    cluster_coe
    print(cluster_coe)
    
    a=a+1
```

```python
#strategy B
G2B=G.copy()



a = 1;
while a<11:
    deg_london =nx.degree_centrality(G2B)
    nx.set_node_attributes(G2B,dict(deg_london),'degree')

    df = pd.DataFrame(index=G2B.nodes())

    df['degree'] = pd.Series(nx.get_node_attributes(G2B, 'degree'))
    df_sorted = df.sort_values(["degree"], ascending=False)
    
    degree_values=[(i[1]['degree']) for i in G2B.nodes(data=True)]
    maxbet=[n for n in G2B.nodes() if G2B.nodes[n]['degree'] == max(degree_values)]
    c1=list()
    c1.append(maxbet[0])
    print(c1)
    G2B.remove_nodes_from(c1)

    global_eff=nx.global_efficiency(G2B)
    global_eff
    print(global_eff)

    cluster_coe=nx.average_clustering(G2B)
    cluster_coe
    print(cluster_coe)
    
    a=a+1
```

```python
#Let us compute the betweenness centrality for the network, without using weights:
bet_london_t=nx.betweenness_centrality(G, normalized=False)
# We can add these values to the nodes attributes:
nx.set_node_attributes(G,bet_london_t,'betweenness_t')

# To ataframe using the nodes as the index
df = pd.DataFrame(index=G.nodes())

df['betweenness_t'] = pd.Series(nx.get_node_attributes(G, 'betweenness_t'))

df_sorted = df.sort_values(["betweenness_t"], ascending=False)
df_sorted[0:10]
```
